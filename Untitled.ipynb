```python
img_requests = []
for imgname in image_filenames:
    with open(imgname, 'rb') as f:
        ctxt = b64encode(f.read()).decode()
        img_requests.append({
                'image': {'content': ctxt},
                'features': [{
                    'type': 'TEXT_DETECTION',
                    'maxResults': 5
                }]
        })
```
